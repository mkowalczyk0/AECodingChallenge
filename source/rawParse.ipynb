```python
import pandas as pd

# Dictionary of dataframe names and their paths
df_paths = {
    'Brands': '../source/cleaned/brands.json',
    'Items': '../source/cleaned/items.json',
    'Receipts': '../source/cleaned/receipts.json',
    'Users': '../source/cleaned/users.json'
}

# Load and print info for each dataframe
for name, path in df_paths.items():
    df = pd.read_json(path)
    print(f"\n{'-'*50}")
    print(f"{name} DataFrame Columns and Types:")
    print(f"{'-'*50}")
    for col, dtype in df.dtypes.items():
        print(f"{col:<30} {dtype}")
```

```python
brandsDF = pd.read_json('../source/cleaned/brands.json')
itemsDF = pd.read_json('../source/cleaned/items.json')
receiptsDF = pd.read_json('../source/cleaned/receipts.json')
usersDF = pd.read_json('../source/cleaned/users.json')


brandsDF['barcode'] = brandsDF['barcode'].astype(str)
itemsDF['barcode'] = itemsDF['barcode'].astype(str)

merged_df = itemsDF.merge(brandsDF, on='barcode', how='left')

merged_df.to_csv('brandedItems.csv')
merged_df.head()
```
